```python
import pandas as pd
import numpy as np

# Step 1: Load the dataset
df = pd.read_csv(r"C:\\Users\\uday2\\Desktop\\python\\DATASET\\new data sets\\Loan-Approval-Prediction.csv")

# Step 2: Check for missing values
print("Missing values before handling:\n", df.isnull().sum())

# Step 3: Drop rows with any missing values (simplest handling)
df_clean = df.dropna().reset_index(drop=True)

# Step 4: Shuffle the data
np.random.seed(42)  # Ensures reproducibility
shuffled_indices = np.random.permutation(len(df_clean))
df_shuffled = df_clean.iloc[shuffled_indices].reset_index(drop=True)

# Step 5: Split data - 70% train, 30% test
split_ratio = 0.7
split_index = int(len(df_shuffled) * split_ratio)
train_df = df_shuffled.iloc[:split_index].reset_index(drop=True)
test_df = df_shuffled.iloc[split_index:].reset_index(drop=True)

# Step 6: Print results
print("\nTraining Set Shape:", train_df.shape)
print("Testing Set Shape:", test_df.shape)

print("\n=== First 5 Rows of Training Set ===")
print(train_df.head())

print("\n=== First 5 Rows of Testing Set ===")
print(test_df.head())
```
